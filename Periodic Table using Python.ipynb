# Code for the information of atoms

```python
import periodictable as pt
#input
Atom=int(input("Enter Atomic Number: "))

features=pt.elements[Atom]

print('Name of Atom: ' ,features.name)
print('Symbol of Atom: ' ,features.symbol)
print('Mass of Atom: ' ,features.mass)
print('Density of Atom: ' ,features.density)

```
